#Questions

<b>Q1.</b> What’s the total RSE by each region and segment?

```python
ans = ps.sqldf("""
---------------------WRITE QUERY BELOW---------------------

SELECT 
    region,
    segment,
    SUM(rse)
FROM rep_data
GROUP BY region, segment

---------------------WRITE QUERY ABOVE---------------------
""")

ans
```

<b>Q2.</b> What’s the average life cycle by each region?

```python
ans = ps.sqldf("""
---------------------WRITE QUERY BELOW---------------------

SELECT 
    region,
    ("deal close date" - "deal start date")
FROM opp_data

---------------------WRITE QUERY ABOVE---------------------
""")

ans
```

<b>Q3.</b> What’s the average deal size by each region? Does classifying them into
small/big or any more categories help us find better trends?

```python
ans = ps.sqldf("""
---------------------WRITE QUERY BELOW---------------------

PRAGMA table_indo(opp_data)

---------------------WRITE QUERY ABOVE---------------------
""")

ans
```

<b>Q4.</b> What’s the conversion rate by each region and segment?

```python
ans = ps.sqldf("""
---------------------WRITE QUERY BELOW---------------------

SELECT 
    p.product_name,
    p.mrp
FROM product_dim AS p
WHERE p.mrp>50000

---------------------WRITE QUERY ABOVE---------------------
""")

ans
```

<b>Q5.</b> What are some insightful trends between different regions/segments and your
recommendations based on these insights?

```python
ans = ps.sqldf("""
---------------------WRITE QUERY BELOW---------------------

SELECT 
    d.pincode,
    COUNT(d.delivery_person_id) AS dp_cnt
FROM delivery_person_dim AS d
GROUP BY d.pincode

---------------------WRITE QUERY ABOVE---------------------
""")

ans
```

<b>Q6.</b> For each Pin code, print the count of orders, sum of total amount paid, average amount paid, maximum amount paid, minimum amount paid for the transactions which were paid by 'cash'. Take only 'buy' order types

```python
ans = ps.sqldf("""
---------------------WRITE QUERY BELOW---------------------

SELECT
    o.delivery_pincode,
    COUNT(o.order_id) AS order_cnt,
    SUM(o.total_amount_paid) AS total_amount_paid,
    AVG(o.total_amount_paid) AS avg_amount_paid,
    MAX(o.total_amount_paid) AS max_amount_paid,
    MIN(o.total_amount_paid) AS min_amount_paid
FROM order_dim as o
WHERE o.payment_type = 'cash'
    AND o.order_type = 'buy'
GROUP BY 
    o.delivery_pincode

---------------------WRITE QUERY ABOVE---------------------
""")

ans
```
